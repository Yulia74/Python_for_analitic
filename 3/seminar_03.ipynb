#### 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU

```python
df['Cpu_Company'] = df['Cpu']  # создала новый признак с нулями
df.head()
```

```python
def get_company_cpy(row: str) -> str:                 # заполнила новый признак через создание функции (у препода через lamda смотри)
    return row.split()[0]

df['Cpu_Company'] = df['Cpu'].apply(get_company_cpy)
df.sample(5)                                         # sample берет выборочно строки
```

#### 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
df['Memory'].value_counts()   #  в начале нужно посмотреть какие есть варианты, чтобы все их учесть в коде
```

```python
# 512GB SSD +  1.0TB Hybrid   -  это самая сложная строка, ее нужно учесть
def mem(row: str) -> int:
    mem_res = 0                 # объявили переменную, которая будет ставиться в новый признак (столбец)
    row_new = row.split()       # разделили строку по split() сделали список из элементов 
    for el in row_new:
        if 'GB' in el:
            mem_res += int(el[:-2])
        elif 'TB' in el:
            mem_res += float(el[:-2])*1024
    return int(mem_res)

df['Memory_Amount'] = df['Memory'].apply(mem)
df.sample(5)
```

#### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)


```python
df['Memory'].value_counts() 
```

```python
# 512GB SSD +  1.0TB Hybrid
def mem_type(row: str) -> str:
    mem_row = []
    row_new = row.split()
    for el in row_new:
        if 'GB' not in el and 'TB' not in el:
            mem_row.append(el)
    return ' '.join(mem_row)

df['Memory_Type'] = df['Memory'].apply(mem_type)
df.sample(10)
```

```python
df['Memory_Type'].value_counts() 
```

#### другой вариант решения задачи

```python
def types(row):
    row = row.split()
    res_str = ''
    for r in row:
        if r.isalpha() and r not in res_str:
            if res_str:
                res_str += ' '
            res_str += r

    return res_str

df['Memory_Type'] = df['Memory'].apply(types)
df.sample(5)
```

#### 1.4 Удалите признаки Memory и ScreenResolution

```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
df.head()
```

#### 4.1 Найдите среднюю стоимость ноутбуков в зависимости от компании производителя
Отсортируйте от меньшей стоимости к большей

```python
grouped_01 = df.groupby('Company') # делает группы, а в них потом агрегации (расчеты)
grouped_01.head()
```

функция agg - словарь {'графа которую надо считать' : 'функция, которую нужно считать'}

#### 4.3 Постройте таблицу с подсчетом количества ноутбуков в данных в зависимости от производителя CPU и ОЗУ

```python
df.pivot_table(index='Cpu_Company',              # сводная таблица показывает зависимости (поэтому кол-во ноутбуков внутри а не в строках и столбцах) 
               columns='Ram', 
               aggfunc='count',
               values='Company',                 # по какой колонке будет считать
               fill_value=0,
               margins=True,
               margins_name='Total')             # fill_value=0 заменяет NaN на 0, если убрать то у Самсунга будет много нанов
```

#### 4.4 Постройте таблицу с подсчетом средней стоимости ноутбуков в данных в зависимости от операционной системы и GB памяти

```python
df.pivot_table(index='OpSys',              # сводная таблица показывает зависимости (поэтому кол-во ноутбуков внутри а не в строках и столбцах) 
               columns='Memory_Amount', 
               aggfunc=('mean'),
               values='Price_euros',                 # по какой колонке будет считать
               fill_value=0,
               margins=True,
               margins_name='Total')   
```
